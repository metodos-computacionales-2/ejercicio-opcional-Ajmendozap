```python
import numpy as np
import matplotlib.pylab as plt

x = np.linspace(0,1,100)
x_med=np.linspace(0,1,50)
def tent(x):
    tent=np.array([])
    for i in range(len(x)):
        if(x[i]<0.5):
            tent= np.append(tent,2*x[i])
        elif(x[i]>=0.5):
            tent= np.append(tent,2*(1-x[i]))
    return tent
def sine(x):
    sine=np.array([])
    for i in range(len(x)):
        sine=np.append(sine,np.sin(np.pi*x[i]))
    return sine 
def logistic(x,L):
    return L*x*(1-x)

tentexponent= np.array([])
tent= tent(x)
for i in range(50):
    tentexponent= np.append(tentexponent,np.log(np.abs((tent[i+1]-tent[i-1]/2))))

sineexponent= np.array([])
sine= sine(x)
for i in range(50):
    sineexponent= np.append(sineexponent,np.log(np.abs((sine[i+1]-sine[i]))))

logisexponent= np.array([])
logis= logistic(x,2)
for i in range(50):
    logisexponent= np.append(logisexponent,np.log(np.abs((logis[i+1]-logis[i]))))
    
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
plt.plot(x,tent)
plt.title("Tent",size=15)
plt.subplot(3,3,2)
plt.plot(x,sine)
plt.title("Sine",size=15)
plt.subplot(3,3,3)
plt.plot(x,logis)
plt.title("Logistic",size=15)
plt.suptitle("Maps",size=20)
plt.savefig("Maps.png")
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
plt.plot(x_med,tentexponent)
plt.title("Tent",size=15)
plt.subplot(3,3,2)
plt.plot(x_med,sineexponent[:50])
plt.title("Sine",size=15)
plt.subplot(3,3,3)
plt.plot(x_med,logisexponent[:50])
plt.title("Logistic",size=15)
plt.suptitle("Lypuanov exponents",size=20)
plt.savefig("Lypuanov exponents.png")
plt.show()
```
